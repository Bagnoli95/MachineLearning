### Métricas en Clasificación

##### Exactitud (Accuracy)
La exactitud mide la proporción de predicciones correctas sobre el total de predicciones. Es una métrica simple y fácil de entender, pero puede ser engañosa si las clases están desbalanceadas.
Fórmula:
Interpretación: Un valor alto de exactitud indica que el modelo está haciendo un buen trabajo en general, pero no garantiza un buen rendimiento en cada clase individualmente.



### Precisión (Precision)
La precisión mide la proporción de verdaderos positivos sobre el total de predicciones positivas (verdaderos positivos + falsos positivos). Indica qué tan precisas son las predicciones positivas del modelo.
Fórmula:
Interpretación: Un valor alto de precisión indica que el modelo tiene pocos falsos positivos.



### Recall (Sensibilidad o Tasa de Verdaderos Positivos)
El recall mide la proporción de verdaderos positivos sobre el total de instancias reales positivas (verdaderos positivos + falsos negativos). Indica qué tan bien el modelo captura las instancias positivas.
Fórmula:
Interpretación: Un valor alto de recall indica que el modelo tiene pocos falsos negativos.

### F1-Score
El F1-Score es la media armónica de la precisión y el recall. Es una métrica balanceada que considera tanto los falsos positivos como los falsos negativos.

### Sensibilidad (Sensitivity) o Recall
La sensibilidad, también conocida como recall, es la proporción de verdaderos positivos entre el total de verdaderos positivos y falsos negativos. Mide la capacidad del modelo para identificar correctamente las instancias positivas.
Fórmula:
Interpretación: Un valor alto de sensibilidad indica que el modelo puede identificar la mayoría de las instancias positivas.

### Especificidad (Specificity)
La especificidad mide la proporción de verdaderos negativos entre el total de verdaderos negativos y falsos positivos. Indica la capacidad del modelo para identificar correctamente las instancias negativas.
Fórmula:
Interpretación: Un valor alto de especificidad indica que el modelo puede identificar la mayoría de las instancias negativas.
